```python
median_steps_per_month = fitbit.groupby(by='Months_encoded')['Steps'].median()
median_steps_wk = fitbit.groupby(by='Work_or_Weekend')['Steps'].median()
```

```python
# MEDIAN STEPS BY MONTH_ENCODED
fig, ax = plt.subplots(2, figsize=(12,4))
ax[0].plot(median_steps_per_month, '-')


# MEDIAN STEPS BY WORK_OR_WEEKEND
ax[1].bar(fitbit['Work_or_Weekend'].unique(), median_steps_wk)
```
